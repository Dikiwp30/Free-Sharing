# Primitive Data Types

Terdapat 4 tipe data dasar yang bisa kita simpan dalam sebuah variable, antara lain:
1. integer 
2. float
3. string
4. boolean 



Tiap tipe data memiliki karakteristiknya sendiri dan digunakan untuk kebutuhan yang berbeda. Untuk mengetahui tipe data yang tersimpan dalam sebuah variable. Kita dapat menggunakan perintah `type()`.

## Integer (int)

Integer adalah data bilangan bulat, baik positif ataupun negatif. Tipe data ini biasanya digunakan pada operasi matematika.

int.PNG

```
nilai = 90
print(nilai)
type(nilai)
```

```
tahun = 2022
print(tahun)
type(tahun)
```

```
suhu = 37
print(suhu)
type(suhu)
```

## Float

Float adalah data bilangan desimal, baik positif ataupun negatif. Sama seperti int, float juga biasanya digunakan pada operasi matematika.


float.PNG

```
gravitasi = 9.8
print(gravitasi)
type(gravitasi)
```

```
phi = 3.14
print(phi)
type(phi)
```

# String (str)

String adalah data yang tersusun atas satu atau lebih karakter (huruf, angka, simbol, dll). String biasanya digunakan pada manipulasi teks.


str.PNG

```
program = "Foundation of AI and Life Skills for Gen Z"
print(program)
type(program)
```

```
referensi = "Dasar pemrograman Python 3"
print(referensi)
type(referensi)
```

# Boolean (bool)

Boolean adalah data yang hanya terdiri dari dua nilai, yakni *True* dan *False*. Boolean biasanya digunakan untuk menentukan nilai kebenaran dari suatu kondisi. 

bool.PNG


```
mahasiswa_aktif = False
type(mahasiswa_aktif)
```

```
mas_ade = 80
mba_farah = 85
nilai_akhir = mas_ade <= mba_farah
print(nilai_akhir)
type(nilai_akhir)
```

# Non-Primitive Data Types

Terdapat 4 tipe data dasar yang bisa kita simpan dalam sebuah variable, antara lain:
1. list
2. tuple
3. set
4. dictionary
5. array



Tiap tipe data memiliki karakteristiknya sendiri dan digunakan untuk kebutuhan yang berbeda. Untuk mengetahui tipe data yang tersimpan dalam sebuah variable. Kita dapat menggunakan perintah `type()`.

# LIST

List adalah tipe data yang paling serbaguna yang tersedia dalam bahasa Python, yang dapat ditulis sebagai daftar nilai yang dipisahkan koma (item) antara tanda kurung siku. Elemen List pada Python tidak harus memiliki tipe data yang sama.

```
akurasi = [0.6, 0.7, 0.8, 0.9]
print(akurasi)
type(akurasi)
```

```
ganjil = [1, 3, 5, 7, 9]
genap = [2, 4, 6, 8]
cacah = ganjil + genap
cacah
```

```
type(cacah)
```

```
cities = ["oslo", "delhi", "washington", "london", "seattle", "paris", "washington", "bekasi", "city"]
cities.index('oslo')
```

```
cities = ['oslo', 'delhi', 'washington', 'london', 'seattle', 'paris', 'washington', 'bekasi', 'city']
cities.reverse() #membalik
cities
```

```
cities.append('brussels') #menambahkan
cities
```

# TUPLE

Tuple adalah jenis tipe data pada Python yang tidak dapat diubah elemennya. Umumnya adalah tuple digunakan untuk data yang bersifat sekali tulis, dan dapat dieksekusi lebih cepat. Tuple didefinisikan dengan kurung dan elemen yang dipisahkan dengan koma.

```
mytuple = ("apple", "banana", "cherry")
print(type(mytuple))
```

```
print(mytuple)
```

# SET

Set adalah kumpulan item bersifat unik dan tanpa urutan (unordered collection). Didefinisikan dengan kurawal dan elemennya dipisahkan dengan koma. Pada Set kita dapat melakukan union dan intersection, sekaligus otomatis melakukan penghapusan data duplikat.

```
thisset = {"apple", "banana", "cherry", "apple", "apple", "banana"}

print(thisset)
```

```
set1 = {"abc", 34, True, 4.0, "male"}

print(set1)
```

# DICTIONARY

Dictionary Python berbeda dengan List ataupun Tuple. Karena setiap urutanya berisi key dan value. Setiap key dipisahkan dari value-nya oleh titik dua (:), item dipisahkan oleh koma, dan semuanya tertutup dalam kurung kurawal. Dictionary kosong tanpa nilai dapat ditulis hanya dengan dua kurung kurawal, seperti ini: {}.

```
profil = {'nama':'diki', 'umur':'20', 'jenis_kelamin':'pria'}
print(profil)
type(profil)
```

```
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964,
  "mobil": ["avanza", "innova", "crv"],
  "nested_dictionary": {
      "halo": "surabaya",
      "nomor": 1
  }
}
print(thisdict)
```

```
type(thisdict)
```

# ARRAY

Tipe Data Array pada Python merupakan variabel yang digunakan untuk menyimpan sekumpulan data dengan tipe yang sama. Jadi misalkan datanya merupakan data numerik, maka semua data yang ada dalam variabel tersebut harus merupakan data numerik. Kita perlu meng-import library NumPy terlebih dahulu untuk bisa membuat dan menggunakan array.

Array dapat dibuat secara langsung melalui <strong>NumPy

```
import numpy as np
```

```
number = np.array([12, 3, 5, 6, 0, -1, 2])
print(number)
type(number)
```

```
huruf = ['Z', 'R', 'A', 'Q', 'W']
huruf = np.array(huruf)
print(huruf)
type(huruf)
```

<h2>If-Else<h2>

```
a = 200
b = 50
if b > a:
  print("b is greater than a")
elif a == b:
  print("a and b are equal")
else:
  print("a is greater than b")
```

<h2>For Statement or Looping<h2>

```
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  print(x)
```

```
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  print(x)
  if x == "banana":
    break
```

```
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  if x == "banana":
    continue
  print(x)
```
